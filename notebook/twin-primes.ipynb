# The Middle Number Conjecture

In 1996, Stephen Wagler communicated a question to Paulo Ribenboim: what is the smallest
middle number that is not the sum of two t-primes? Ribenboim asked Harvey Dubner to investigate.
Dubner's paper (1999–2000) is the earliest known published treatment; he formulated four
conjectures in response. Wagler's question is closest in spirit to Conjecture 2 (see below).

**Conjecture 1** (Dubner, 1999): *Every middle number greater than 6 is the sum of two middle numbers.*

A "middle number" is a multiple of 6 sandwiched between a pair of twin primes: if m is in A002822
(i.e. 6m−1 and 6m+1 are both prime), then 6m is the corresponding middle number. A002822 records
the cofactor m, not the middle number itself. A t-prime is a prime which has a twin.

Conjecture 1, translated into A002822 terms, states: for every m in A002822 with 6m > 6,
there exist a, b in A002822 such that 6m = 6a + 6b, i.e. m = a + b.

Dubner verified Conjecture 1 up to 2×10¹⁰, finding that the following multiples of 6 are not
sums of two middle numbers: 96, 402, 516, 786, 906, 1116, 1146, 1266, 1356, 3246, 4206.
None of these are middle numbers themselves, so Conjecture 1 (which applies only to middle
numbers) was not contradicted. The largest non-representable multiple of 6 is 4206, leading to:

**Conjecture 3** (Dubner): *Every multiple of 6 greater than 4206 is the sum of two middle numbers.*

Dubner also states two further conjectures involving t-primes directly:

**Conjecture 2** (Dubner): *Every middle number greater than 6 is the sum of two t-primes.*

**Conjecture 4a** (Dubner): *Every even number greater than 4208 is the sum of two t-primes.*

Conjecture 4a is a strengthening of Goldbach's conjecture requiring both summands to be t-primes
(primes with a twin). It is closely related to our Goldbach-Twin Conjecture below.

We compute, for each m in A002822, the list of pairs (a, b) in A002822 such that a + b = m.

# Classification of Integers via 6ab ± a ± b

## The Four Factoring Identities

For all integers a, b, the four sign combinations of `6ab ± a ± b` satisfy:

| Form | Identity |
|------|----------|
| 6ab + a + b | 6(6ab + a + b) + 1 = (6a+1)(6b+1) |
| 6ab − a − b | 6(6ab − a − b) + 1 = (6a−1)(6b−1) |
| 6ab + a − b | 6(6ab + a − b) − 1 = (6a−1)(6b+1) |
| 6ab − a + b | 6(6ab − a + b) − 1 = (6a+1)(6b−1) |

Each identity is verified by expanding the right-hand side. Conversely, if 6N+1 = PQ
with P = 6a±1 and Q = 6b±1 (a,b ≥ 1), the identities give an explicit (a,b) with
6ab ± a ± b = N. Likewise for 6N−1.

## Theorem (U ∈ {0, ±1} suffices)

Every integer N ≥ 3 is representable as:

```
N = 6ab ± a ± b + U,    a, b ≥ 1,  U ∈ {−1, 0, 1}
```

In particular, any representation `6ab ± a ± b + D` with D ∈ {2,3,4,5} can always be
replaced by a different (a', b') with U' ∈ {−1, 0, 1}.

### Proof

**Case 1: at least one of 6N−1, 6N+1 is composite.**

Every integer greater than 1 that is not divisible by 2 or 3 is of the form 6k±1,
and so are all its factors. If 6N+1 = PQ with P = 6a+ε₁, Q = 6b+ε₂ (ε₁,ε₂ ∈ {±1},
a,b ≥ 1), the factoring identities give 6ab + ε₁a + ε₂b = N directly — i.e. U = 0.
The same applies if 6N−1 is composite.

**Case 2: both 6N−1 and 6N+1 are prime.**

Consider the six numbers of the form 6k±1 for k ∈ {N−1, N, N+1}:

```
6(N−1)−1,  6(N−1)+1,  6N−1,  6N+1,  6(N+1)−1,  6(N+1)+1
```

Among any five consecutive integers, one is divisible by 5. The six numbers above include
five consecutive odd non-multiples of 3, which among themselves must contain a multiple of 5.
Since 6N−1 and 6N+1 are assumed prime and greater than 5 (for N ≥ 2), neither of them is
that multiple. Therefore the multiple of 5 is one of 6(N−1)±1 or 6(N+1)±1.

A number of the form 6k±1 that is divisible by 5 equals 5 · (something of the form 6j±1),
and since it exceeds 5 (for N ≥ 2) it is composite. Composite 6k±1 means k falls into
Case 1 — so either N−1 or N+1 has a U=0 representation 6ab ± a ± b = N∓1, giving
N = 6ab ± a ± b ± 1, i.e. U = ±1. ∎

### Computational verification

Verified for all N in [3, 1000]: no integer requires |U| > 1 when a, b ≥ 1.

### Why the syntactic argument works only for D = 5

One might hope to handle D ∈ {2,3,4,5} by a purely syntactic rewrite — replacing the
index (a,b) rather than invoking primality. This works for D=5 but not for D∈{2,3,4},
for a fundamental modular reason.

**D = 5 (works):** Note that 5 ≡ −1 (mod 6), so 6n+5 = 6(n+1)−1. Any number of the
form 6n+5 is already of the form 6k−1 (with k = n+1), which sits inside the 6k±1 family
that the factoring identities act on. The rewrite is a free index shift — no new (a,b)
arithmetic is required, just a relabelling.

**D ∈ {2, 3, 4} (fail):** These residues mod 6 are:

| D | 6n+D mod 6 | Divisible by |
|---|-----------|--------------|
| 2 | 2         | 2            |
| 3 | 3         | 3            |
| 4 | 4         | 2            |

Each of 6n+2, 6n+3, 6n+4 is divisible by 2 or 3, so it is *never* of the form 6k±1.
The factoring identities only produce values of the form 6k±1, and no syntactic index
shift can move a number out of its residue class mod 6. There is therefore no syntactic
rewrite into the 6k±1 family for D∈{2,3,4} — the factoring/primality argument is
genuinely required.

In short: D=5 is the unique value in {2,3,4,5} that is ≡ ±1 (mod 6), and it is exactly
this residue membership that makes the syntactic argument possible.

## Connection to A002822

The theorem is purely elementary — it makes no reference to twin primes or A002822.
A002822 emerges naturally as a consequence:

- **N requires U = 0**: at least one of 6N±1 is composite, so N ∉ A002822. The (a,b)
  witness from Case 1 is exactly what the `ab` function in this notebook computes.

- **N requires U = ±1**: both 6N−1 and 6N+1 are prime, so N ∈ A002822. The U=0
  representation exists only for N−1 or N+1, not for N itself.

**A002822 is therefore exactly the set of N ≥ 3 for which U = 0 fails** — i.e. where
no non-trivial factoring of 6N±1 exists — **and U = ±1 is always sufficient as a fallback.**

# Congruence Constraints on Middle Number Decompositions

We derive necessary congruence conditions on `m_0 + m_1 = m_2` when all three lie in A002822.

## Residues of A002822 elements mod 5

Since `6 ≡ 1 (mod 5)`, we have `6k ≡ k (mod 5)`, so:

| k mod 5 | 6k−1 mod 5 | 6k+1 mod 5 |
|---------|-----------|-----------|
| 0 | 4 | 1 |
| 1 | **0** | 2 |
| 2 | 1 | 3 |
| 3 | 2 | 4 |
| 4 | 3 | **0** |

For `m ∈ A002822` with `m > 1`, both `6m−1` and `6m+1` are prime and greater than 5,
so neither is divisible by 5. Reading the table:

- `6m−1 ≡ 0 (mod 5)` iff `m ≡ 1 (mod 5)` — **forbidden**
- `6m+1 ≡ 0 (mod 5)` iff `m ≡ 4 (mod 5)` — **forbidden**

Therefore, for all `m ∈ A002822` with `m > 1`:

> **`m mod 5 ∈ {0, 2, 3}`**

(The single exception is `m = 1`, where `6·1−1 = 5` is itself prime.)

## Forbidden pairs in middle number decompositions

For `m_0 + m_1 = m_2` with all three in A002822 (and all `> 1`), each residue `r_i = m_i mod 5`
lies in `{0, 2, 3}`, and `r_0 + r_1 ≡ r_2 (mod 5)` must also lie in `{0, 2, 3}`.

Checking all unordered pairs `{r_0, r_1} ⊆ {0, 2, 3}`:

| r_0 | r_1 | (r_0 + r_1) mod 5 | Valid r_2? |
|-----|-----|-------------------|-----------|
| 0 | 0 | 0 | ✓ |
| 0 | 2 | 2 | ✓ |
| 0 | 3 | 3 | ✓ |
| 2 | 2 | 4 | **✗** |
| 2 | 3 | 0 | ✓ |
| 3 | 3 | 1 | **✗** |

**Two pairs are forbidden:**
- `m_0 ≡ m_1 ≡ 2 (mod 5)`: their sum ≡ 4 (mod 5), which no A002822 element (> 1) can have.
- `m_0 ≡ m_1 ≡ 3 (mod 5)`: their sum ≡ 1 (mod 5), which no A002822 element (> 1) can have.

All other residue combinations are permitted (and observed computationally).

## Symmetry observation

Note that 2 and 3 are additive inverses mod 5 (2 + 3 = 5 ≡ 0). The forbidden pairs are
exactly the two cases where both summands share the same non-zero residue from {2, 3}.
Equivalently: a middle number decomposition `m_0 + m_1 = m_2` is impossible if
`m_0 ≡ m_1 ≢ 0 (mod 5)` and `m_0 + m_1 ≡ 0 (mod 5)` is not the outcome —
more precisely, if `2m_0 ≡ 4` or `2m_0 ≡ 1 (mod 5)`, i.e. `m_0 ≡ 2` or `m_0 ≡ 3 (mod 5)`
and `m_1 = m_0` in residue class.

# The Bridging Conjecture

**Bridging Conjecture**: *For all K, there exist a, b, k in A002822 such that a < K, b < K, and k = a + b > K.*

That is: no matter how large K is, one can always find two elements of A002822 below K whose sum
is also in A002822 and exceeds K. In middle number terms: there exist middle numbers 6a and 6b,
both below 6K, whose sum 6k = 6a + 6b is also a middle number and exceeds 6K.

*Note: the current author suspects that Wagler, Dubner, or Zwillinger have already stated an
equivalent conjecture, but has not yet located the precise reference.*

## Relationship to Dubner's Conjecture 3

Dubner's paper states:

**Conjecture 3 (Dubner)**: *Every sufficiently large number divisible by 6 is the sum of two middle numbers.*

More precisely: every multiple of 6 greater than 4206 is the sum of two middle numbers
(i.e. two elements of {6m : m ∈ A002822}).

Conjecture 3 (Dubner) is strictly stronger than Conjecture 1 (Dubner), which only requires
middle numbers themselves to decompose — Conjecture 3 (Dubner) requires *all* sufficiently
large multiples of 6 to decompose, whether or not they are middle numbers.

**Conjecture 3 (Dubner) implies the Bridging Conjecture:** Assume Conjecture 3 (Dubner) holds.
Given any K, let k be the smallest element of A002822 greater than K — such k exists assuming
the Twin Prime Conjecture. Since 6k is a multiple of 6 greater than 4206 (for large enough K),
Conjecture 3 (Dubner) guarantees 6k = 6a + 6b for some a, b ∈ A002822, i.e. k = a + b. Since
k is the *smallest* element of A002822 greater than K, any element of A002822 less than k is
≤ K, so a, b ≤ K. Since a + b = k > K they cannot both equal K, so a, b < K. Thus a, b, k
witness the Bridging Conjecture. ∎

## Relationship to Conjecture 1 (Dubner) and the Twin Prime Conjecture

Conjecture 1 (Dubner) and the Twin Prime Conjecture together imply the Bridging Conjecture:

**Twin Primes ∧ Conjecture 1 → Bridging Conjecture:**
Assume both hold. Given any K, since A002822 is infinite (Twin Primes), let k be the smallest
element of A002822 greater than K. By Conjecture 1, k = a + b for some a, b in A002822 with
a, b < k. Since k is the *smallest* element of A002822 greater than K, any element of A002822
less than k must be ≤ K. Therefore a ≤ K and b ≤ K, and since a + b = k > K they cannot both
equal K, so a, b < K. Thus a, b, k satisfy the Bridging Conjecture. ∎

Conversely, the Bridging Conjecture implies the Twin Prime Conjecture:

**Bridging Conjecture → Twin Primes:**
For any K, the Bridging Conjecture gives an element k = a + b of A002822 with k > K. Since K was
arbitrary, A002822 is infinite — i.e. there are infinitely many twin prime pairs. ∎

Therefore, assuming Conjecture 1 (Dubner), the Bridging Conjecture and the Twin Prime
Conjecture are equivalent:

> Conjecture 1 (Dubner) → ( Bridging Conjecture ↔ Twin Prime Conjecture )

Conjecture 1 itself is independent — it could hold for a finite set of elements of A002822
(in which case the Twin Prime Conjecture is false and the Bridging Conjecture is false), or fail
for an infinite one (in which case the Twin Prime Conjecture could still be true while the
Bridging Conjecture remains open).

# Gap Bound and Density Implications of the Bridging Conjecture

## The gap bound

The Bridging Conjecture gives, for any K, elements a, b, k in A002822 with a < K, b < K,
and k = a + b > K. Since a < K and b < K:

```
k = a + b < K + K = 2K
```

So: **for every K, there exists k in A002822 with K < k < 2K**.

This does not require Conjecture 1 — it is a direct consequence of the Bridging
Conjecture's statement alone.

Equivalently: if m_n < m_{n+1} are consecutive elements of A002822, then m_{n+1} < 2·m_n.
No gap between consecutive elements can exceed the smaller element.

## Density lower bound

Since every interval (m, 2m) contains at least one element of A002822, starting from
m_1 = 1 and applying the gap bound inductively:

```
m_1 = 1,  m_2 < 2,  m_3 < 4,  m_4 < 8,  ...,  m_n < 2^(n-1)
```

Inverting: if m_n ≤ x, then n ≥ log₂(x) + 1. So:

```
π₂(x) = |{m ∈ A002822 : m ≤ x}| ≥ ⌊log₂(x)⌋
```

This is a conditional lower bound: **assuming the Bridging Conjecture, π₂(x) ≫ log x**.

This is far weaker than the conjectured truth (roughly x / log²x by Hardy–Littlewood
heuristics), but it is a rigorous consequence of the Bridging Conjecture alone, with no
appeal to analytic number theory.

## What Conjecture 1 adds

The gap bound and density lower bound follow from the Bridging Conjecture alone. Conjecture 1
(Dubner) says every element of A002822 beyond 1 decomposes as a sum of two smaller elements —
not just the specific k produced by the Bridging Conjecture, but all of them. Note that sums of pairs from A002822 need not land in A002822 — many land in A067611 instead.
What Conjecture 1 asserts is that every element of A002822 beyond 1 is reachable as a sum of
two smaller elements of A002822; no element is an additive dead-end. This is a non-trivial
demand on the self-sumset structure of A002822, and is a stronger statement than the Bridging
Conjecture alone.

# No three consecutive terms in A002822

**Observation** (Jonathan Seymour, May 14 2026):

No three terms of A002822 greater than 3 are consecutive. Otherwise there would exist six integers
6m−1, 6m+1, 6m+5, 6m+7, 6m+11, 6m+13 that are all prime. One of these must be
divisible by 5, which is impossible for primes greater than 5. (The exception {1, 2, 3} corresponds
to the primes {5, 7, 11, 13, 17, 19}, where 5 itself is prime rather than composite.)

Hence every term of A002822 greater than 3 is necessarily of the form 6ab ± a ± b ± 1. This condition
is necessary but not sufficient for membership in the sequence.

````python
def _prime_sieve(limit):
    sieve = bytearray([1]) * (limit + 1)
    sieve[0] = sieve[1] = 0
    for i in range(2, int(limit**0.5) + 1):
        if sieve[i]:
            sieve[i*i::i] = bytearray(len(sieve[i*i::i]))
    return sieve

def A002822(seq, sieve):
    for m in seq:
        if sieve[6*m-1] and sieve[6*m+1]:
            yield m

def A067611(seq, sieve):
    for m in seq:
        if not (sieve[6*m-1] and sieve[6*m+1]):
            yield m

def ab(n):
    for b in range(1, n):
        for a in range(1, b+1):
            if 6*a*b-a-b >= n:
                break
            yield {"a": -a, "b": -b, "c": 0, "m": 6*a*b-a-b}
            yield {"a": +a, "b": -b, "c": 0, "m": 6*a*b+a-b}
            yield {"a": -a, "b": +b, "c": 0, "m": 6*a*b-a+b}
            yield {"a": +a, "b": +b, "c": 0, "m": 6*a*b+a+b}

def ab_dict(n):
    d={}
    for ab_ in ab(n):
        m = ab_["m"]
        if not m in d:
            d[m] = []
        d[m].append(ab_)
    return d

def a002822_augmented(n):
    sieve = _prime_sieve(6*n + 1)
    list_2822 = list(A002822(range(1, n+1), sieve))
    set_67611 = set(A067611(range(1, n+1), sieve))
    d = ab_dict(n+1)
    def augment(e, l, c):
        if c > 0:
            out=e|{"m": l, "c": c, "x": e["a"], "y": e["b"]}
        else:
            out=e|{"m": l, "c": c, "x": e["b"], "y": e["a"]}
        out["l"]=6*l-1
        out["u"]=6*l+1
        return out

    for l in list_2822:
        if l-1 in set_67611:
            ab = d[l-1]
            for e in ab:
                yield augment(e, l, 1)
        if l+1 in set_67611:
            ab = d[l+1]
            for e in ab:
                yield augment(e, l, -1)
````

````python
def plot_gap_ratio_bound(N=5000):
    """Plot m_{n+1}/m_n for A002822 elements to verify the gap bound m_{n+1} < 2*m_n
    implied by the Bridging Conjecture. All ratios should be strictly less than 2.
    """
    import matplotlib.pyplot as plt
    import numpy as np

    sieve = _prime_sieve(6*N + 1)
    seq = [m for m in range(1, N+1) if sieve[6*m-1] and sieve[6*m+1]]

    ratios = [seq[i+1] / seq[i] for i in range(len(seq)-1)]
    indices = list(range(1, len(ratios)+1))

    # Rolling mean trend line
    window = max(1, len(ratios) // 50)
    trend = np.convolve(ratios, np.ones(window)/window, mode='valid')
    trend_x = indices[window//2: window//2 + len(trend)]

    fig, ax = plt.subplots(figsize=(12, 4))
    ax.scatter(indices, ratios, s=1, color='steelblue', alpha=0.3, label='m_{n+1}/m_n')
    ax.plot(trend_x, trend, color='navy', linewidth=1.2, label=f'Rolling mean (window={window})')
    ax.axhline(y=2.0, color='red', linestyle='--', linewidth=1, label='Bound: ratio = 2')
    ax.axhline(y=1.0, color='gray', linestyle=':', linewidth=0.8)
    ax.set_xlabel('n')
    ax.set_ylabel('m_{n+1} / m_n')
    ax.set_title(f'Consecutive ratio m_{{n+1}}/m_n for A002822 (up to m={seq[-1]})')
    ax.legend()
    ax.set_ylim(0.9, 2.1)
    plt.tight_layout()
    plt.show()

    max_ratio = max(ratios)
    print(f"Maximum consecutive ratio: {max_ratio:.6f}  (bound: < 2.0)")
    print(f"All ratios < 2: {all(r < 2 for r in ratios)}")
    print(f"Elements checked: {len(seq)}, largest: {seq[-1]}")

plot_gap_ratio_bound(N=5000)
````

````python
def verify_wagler_mod5(n=10000):
    """Verify the mod-5 congruence constraints on middle number decompositions.

    Confirms:
    - A002822 elements > 1 have residues in {0, 2, 3} mod 5 only.
    - No middle number decomposition m0 + m1 = m2 has (m0 mod 5, m1 mod 5) in {(2,2),(3,3)}.
    - All other residue combinations do occur.
    """
    from collections import Counter

    sieve = _prime_sieve(6*n + 1)
    centers = [m for m in range(1, n+1) if sieve[6*m-1] and sieve[6*m+1]]
    center_set = set(centers)

    # Residues mod 5 (excluding m=1 exception)
    residues = Counter(m % 5 for m in centers if m > 1)
    print(f"Residues mod 5 for A002822 elements > 1 (n={n}):")
    for r in range(5):
        mark = "  <- forbidden" if r in {1, 4} else ""
        print(f"  r={r}: {residues[r]} elements{mark}")

    # Check all middle number decompositions
    pair_counts = Counter()
    forbidden_found = []
    for m2 in centers:
        for m0 in centers:
            if m0 >= m2:
                break
            m1 = m2 - m0
            if m1 in center_set and m1 >= m0:
                r0, r1 = m0 % 5, m1 % 5
                pair_counts[(min(r0,r1), max(r0,r1))] += 1
                if (r0, r1) in {(2, 2), (3, 3)}:
                    forbidden_found.append((m0, m1, m2))

    print(f"\nUnordered residue pairs (r0, r1) in middle number decompositions:")
    for pair in sorted(pair_counts):
        r0, r1 = pair
        status = "FORBIDDEN (never occurs)" if pair in {(2,2),(3,3)} else "ok"
        print(f"  ({r0},{r1}): {pair_counts[pair]} decompositions  [{status}]")

    # Also show the two forbidden pairs explicitly with count 0
    for pair in [(2,2),(3,3)]:
        if pair not in pair_counts:
            print(f"  {pair}: 0 decompositions  [FORBIDDEN — confirmed absent]")

    if forbidden_found:
        print(f"\nFORBIDDEN PAIRS FOUND: {forbidden_found[:5]}")
    else:
        print(f"\nConfirmed: no (2,2) or (3,3) residue pairs appear in any middle number decomposition.")

verify_wagler_mod5()
````

````python
def generate_from_seeds(seeds, N):
    """Starting from a seed set, repeatedly take the largest unprocessed member m_max,
    add it to each smaller member, and admit the sum if its residue mod 5 is in {0,2,3}.
    Continue until all members exceed N.

    Returns the generated set M.
    """
    M = sorted(seeds)
    M_set = set(M)
    processed = set()

    while True:
        remaining = [m for m in M if m not in processed and m <= N]
        if not remaining:
            break
        m_max = max(remaining)
        processed.add(m_max)
        for m_i in [m for m in M if m < m_max]:
            s = m_max + m_i
            if s <= N and s % 5 in {0, 2, 3} and s not in M_set:
                M_set.add(s)
                M = sorted(M_set)

    return M_set

def compare_to_a002822(seeds, N=300):
    sieve = _prime_sieve(6*N + 1)
    a002822 = set(m for m in range(1, N+1) if sieve[6*m-1] and sieve[6*m+1])

    M = generate_from_seeds(seeds, N)

    in_M_not_a = sorted(M - a002822)
    in_a_not_M = sorted(a002822 - M)

    print(f"Seeds: {sorted(seeds)},  N={N}")
    print(f"  Generated |M| = {len(M)},  |A002822| = {len(a002822)}")
    print(f"  In M but not A002822 ({len(in_M_not_a)}): {in_M_not_a[:20]}{'...' if len(in_M_not_a)>20 else ''}")
    print(f"  In A002822 but not M ({len(in_a_not_M)}): {in_a_not_M[:20]}{'...' if len(in_a_not_M)>20 else ''}")

    # For elements in A002822 but not M, show their middle number decompositions
    if in_a_not_M:
        print(f"\n  A002822 \\ M — middle number decompositions within A002822:")
        for m in in_a_not_M[:15]:
            decomps = [(a, m-a) for a in sorted(a002822) if a < m and (m-a) in a002822]
            print(f"    m={m} (mod5={m%5}): {decomps[:4]}{'...' if len(decomps)>4 else ''}")

compare_to_a002822(seeds={2, 3}, N=300)
````

# The Growth Rule as a Mod-5 Sieve

The process of growing M from seeds {1, 2, 3} by admitting sums whose residue mod 5 lies
in {0, 2, 3} is precisely the **p=5 layer of the sieve of Eratosthenes** applied to the
family {6m±1}.

## What the mod-5 sieve does

For each prime p, the sieve of Eratosthenes eliminates the residue classes of m for which
p divides 6m±1. For p=5, since 6 ≡ 1 (mod 5), this eliminates:

- m ≡ 1 (mod 5): makes 5 | 6m−1
- m ≡ 4 (mod 5): makes 5 | 6m+1

leaving exactly m ≡ {0, 2, 3} (mod 5) as survivors. The growth rule encodes this filter
directly — admitting a sum only when its residue lies in {0, 2, 3} is the same as applying
the p=5 sieve to each candidate.

## Why it cannot capture larger primes

The sieve layers for p=7, 11, 13, … live in ℤ/7ℤ, ℤ/11ℤ, ℤ/13ℤ, … respectively. By the
Chinese Remainder Theorem, these are independent of ℤ/5ℤ: a condition on m mod 5 carries
no information about m mod 7. Concretely:

- The condition 7 | 6m−1 means m ≡ 6 (mod 7). This residue class is spread uniformly
  across all five residues mod 5, so no mod-5 filter can detect or eliminate it.
- The same holds for every prime p ≥ 7: its two forbidden residue classes mod p are
  invisible to arithmetic in ℤ/5ℤ.

The growth rule is closed under mod-5 arithmetic — sums of elements of M are filtered in
ℤ/5ℤ, producing a set closed under this filtered addition. This is exactly what a mod-5
sieve produces, and it is the *only* sieve such an operation can implement.

## The excess as sieve incompleteness

A002822 is the intersection over all primes p of the sets S_p of m not eliminated at p:

```
A002822 = ⋂_p  S_p
```

The growth rule implements only S_5 (restricted to sums reachable from the seeds). The
excess M \ A002822 is the set of elements that survive the mod-5 filter but are eliminated
by some larger prime — i.e. elements of M ∩ (S̄_7 ∪ S̄_11 ∪ S̄_13 ∪ …). This excess is
unavoidable: no sum-closure operation on residues mod 5 can approximate a sieve at any
other prime, because the relevant residue classes are algebraically independent.

# Product–Sum Identity for Middle Number Decompositions

## Setup

Every integer m has a representation:

```
m = 6ab + εa + δb + u,    a, b ≥ 1,  ε, δ ∈ {±1},  u ∈ {−1, +1}
```

(The Classification Theorem shows U ∈ {−1, 0, 1} suffices; here we absorb the U=0 case
into U=±1 by noting that if U=0 then m ∉ A002822 and we may shift to a neighbour.)

Label three such representations with subscripts i, j, k:

```
mᵢ = 6aᵢbᵢ + εᵢaᵢ + δᵢbᵢ + uᵢ
mⱼ = 6aⱼbⱼ + εⱼaⱼ + δⱼbⱼ + uⱼ
mₖ = 6aₖbₖ + εₖaₖ + δₖbₖ + uₖ
```

with uᵢ, uⱼ, uₖ ∈ {−1, +1}.

## The identity

Imposing mₖ = mᵢ + mⱼ and grouping product terms on the left, additive terms on the right:

```
6(aₖbₖ − aᵢbᵢ − aⱼbⱼ) = εᵢaᵢ + δᵢbᵢ + εⱼaⱼ + δⱼbⱼ − εₖaₖ − δₖbₖ + uᵢ + uⱼ − uₖ
```

## The parity constraint

Since uᵢ, uⱼ, uₖ ∈ {−1, +1}, the quantity uᵢ + uⱼ − uₖ ∈ {−3, −1, +1, +3} is always
odd. The left-hand side is divisible by 6, so the right-hand side must also be divisible
by 6. In particular it must be even, so the signed linear combination

```
S = εᵢaᵢ + δᵢbᵢ + εⱼaⱼ + δⱼbⱼ − εₖaₖ − δₖbₖ
```

must be odd (odd + odd = even, as required). This is the only parity constraint: S must
be odd, which is equivalent to S + (uᵢ + uⱼ − uₖ) being divisible by 6.

## The deficit is small

Rearranging:

```
aₖbₖ − aᵢbᵢ − aⱼbⱼ = (S + uᵢ + uⱼ − uₖ) / 6
```

The right-hand side is O(max(aᵢ, bᵢ, aⱼ, bⱼ, aₖ, bₖ)), while the left-hand side
involves products. Since mₖ ≈ 6aₖbₖ and mₖ = mᵢ + mⱼ ≈ 6aᵢbᵢ + 6aⱼbⱼ, the leading
behaviour is aₖbₖ ≈ aᵢbᵢ + aⱼbⱼ. The identity makes this precise: the deficit

```
aₖbₖ − aᵢbᵢ − aⱼbⱼ = O(max(a, b))
```

is linear in the parameters, much smaller than the quadratic scale of the products
themselves. In other words, the product structure is nearly additive, with a correction
that is small relative to m.

## Connection to Conjectures 1, 2, and 3 (Dubner)

For m ∈ A002822, write p₋₁(m) = 6m−1 and p₁(m) = 6m+1 for the associated twin primes.
If mᵢ, mⱼ ∈ A002822 and mₖ = mᵢ + mⱼ, then regardless of whether mₖ ∈ A002822:

```
6mₖ = 6(mᵢ + mⱼ) = p₁(mᵢ) + p₋₁(mⱼ)  =  p₋₁(mᵢ) + p₁(mⱼ)
```

So whenever such a pair exists, 6mₖ is a sum of two t-primes. Since mₖ is either in
A002822 or A067611, the pair (when it exists) witnesses different conjectures in each case:

- **mₖ ∈ A002822** (6mₖ is a middle number): a decomposition mₖ = mᵢ + mⱼ is a Wagler
  decomposition (Conjecture 1), and 6mₖ = p₁(mᵢ) + p₋₁(mⱼ) simultaneously witnesses
  **Conjecture 2 (Dubner)** — every middle number is a sum of two t-primes. Thus
  **Conjecture 1 (Dubner) implies Conjecture 2 (Dubner)** as a free corollary.

- **mₖ ∈ A067611** (6mₖ is not a middle number): a decomposition mₖ = mᵢ + mⱼ, if one
  exists, witnesses **Conjecture 3 (Dubner)** for this particular multiple of 6. But the
  existence of such a pair is exactly what Conjecture 3 (Dubner) asserts — the identity
  only says that *having* the pair is sufficient, not that the pair exists.

## Expressions for 6mₖ ± 1

Adding and subtracting 1 from 6mₖ = p₁(mᵢ) + p₋₁(mⱼ):

```
6mₖ + 1 = p₁(mᵢ) + p₋₁(mⱼ) + 1 = p₁(mᵢ) + p₁(mⱼ) − 1
6mₖ − 1 = p₁(mᵢ) + p₋₁(mⱼ) − 1 = p₋₁(mᵢ) + p₋₁(mⱼ) + 1
```

These express 6mₖ±1 as sums of two t-primes shifted by ±1. Whether 6mₖ±1 is prime
depends on mₖ ∈ A002822 — but even then, primality of these shifted sums is not implied
by the decomposition alone. We record the algebra for completeness; no stronger conclusion
is available from the identity.

## D as the complement of the sumset of A002822

[A243956](https://oeis.org/A243956) is defined as the complement of { i + j : i, j ∈ A002822 } with respect to ℕ. That is:

```
D = N \ { i + j : i, j ∈ A002822 }
  = {1, 16, 67, 86, 131, 151, 186, 191, 211, 226, 541, 701}  (up to 2×10¹⁰)
```

By definition, { i + j : i, j ∈ A002822 } and D partition ℕ≥1 — this is a tautology,
not a conjecture. Conjecture 3 (Dubner) is the separate assertion that D is finite.

Several structural observations about D:

- **1 ∈ A002822** (since 5 and 7 are prime) but 1 ∈ D, because no two elements of
  A002822 sum to 1. It is the only element of D that is also in A002822.
- Every other element of D lies in A067611 (the complement of A002822): none of the
  eleven exceptions beyond 1 is a twin-prime cofactor.
- Once A002822 is given, D is determined entirely by its additive structure —
  no further reference to primality is needed.
- Dubner found D top-down by searching for multiples of 6 not expressible as a sum
  of two middle numbers. The sumset approach finds D bottom-up: enumerate all sums
  i + j with i, j ∈ A002822 and take the complement.
- A new exception beyond 701 would falsify Conjecture 3 (Dubner) but would merely
  enlarge D in the definition above, leaving the partition intact.
- The size and composition of D does not by itself determine the truth of the Bridging
  Conjecture or the Twin Prime Conjecture. D could be infinite and both conjectures
  could still hold — provided the sumset { i+j : i,j ∈ A002822 } remains unbounded.
  What matters is not the size of D but whether A002822 itself is infinite. Note,
  however, that the Bridging Conjecture does place an indirect constraint on D: it
  requires infinitely many elements of A002822 to be reachable as sums of pairs of
  smaller elements of A002822, so D cannot eventually absorb all positive integers.

## Relationship between Conjecture 3 and Conjecture 4a exception sets

The exception set for Conjecture 4a (Dubner) — even numbers not expressible as a sum of
two t-primes, recorded as OEIS A007534 — is:

```
2, 4, 94, 96, 98, 400, 402, 404, 514, 516, 518, 784, 786, 788, 904, 906, 908,
1114, 1116, 1118, 1144, 1146, 1148, 1264, 1266, 1268, 1354, 1356, 1358,
3244, 3246, 3248, 4204, 4206, 4208
```

These cluster in triples {6m−2, 6m, 6m+2} around each Conjecture 3 exception 6m. This
is exactly what the identity predicts: since 6m = p₁(mᵢ) + p₋₁(mⱼ), the natural
candidates for 6m±2 are p₁(mᵢ) + p₁(mⱼ) − 2 = 6m and p₋₁(mᵢ) + p₋₁(mⱼ) + 2 = 6m, so
when no decomposition exists for 6m, the even numbers 6m−2 and 6m+2 also lose their most
natural t-prime witnesses. The two small exceptions 2 and 4 are simply too small to have
any t-prime representation at all.

````python
def verify_dubner_conjecture3(N=10000):
    """Verify Dubner's Conjecture 3 (set form): the positive integers with no
    representation as a sum i + j with i, j in A002822 are exactly:
      {1, 16, 67, 86, 131, 151, 186, 191, 211, 226, 541, 701}

    Checks all positive integers up to N and reports any exceptions beyond 701.
    """
    A243956 = {1, 16, 67, 86, 131, 151, 186, 191, 211, 226, 541, 701}

    sieve = _prime_sieve(6*N + 1)
    a002822 = sorted(m for m in range(1, N+1) if sieve[6*m-1] and sieve[6*m+1])
    a002822_set = set(a002822)

    # Build sumset: all i+j with i, j in A002822, up to N
    sumset = set()
    for i in a002822:
        for j in a002822:
            s = i + j
            if s > N:
                break
            sumset.add(s)

    exceptions = sorted(n for n in range(1, N+1) if n not in sumset)

    known = sorted(A243956)
    unexpected = [n for n in exceptions if n not in A243956]
    missing = [n for n in A243956 if n <= N and n not in exceptions]

    print(f"Checked n in [1, {N}]:")
    print(f"  Exceptions (not in sumset of A002822): {exceptions}")
    print(f"  Known A243956:                         {known}")
    print(f"  Unexpected exceptions beyond A243956:  {unexpected if unexpected else 'none'}")
    print(f"  A243956 members missing from exceptions: {missing if missing else 'none'}")
    if not unexpected and not missing:
        print(f"  Conjecture 3 verified up to N={N}.")

verify_dubner_conjecture3(N=10000)
````

# Characterising the Excess

With seeds `{1, 2, 3}`, the generated set M fully contains A002822 but also includes many
elements not in A002822. The excess M \ A002822 is simply M ∩ A067611 — elements where at
least one of 6m−1, 6m+1 is composite.

## Why the excess cannot be characterised more tightly

The mod-5 gate rules out exactly two residue classes (m ≡ 1 and m ≡ 4 mod 5), corresponding
to the prime p=5 dividing one of 6m±1. But it says nothing about any other prime p ≡ ±1 (mod 6).
For each such prime p, there are two residue classes of m (mod p) for which p | 6m±1:

| prime p | m mod p where p \| 6m−1 | m mod p where p \| 6m+1 |
|---------|------------------------|------------------------|
| 7  | 6 | 1 |
| 11 | 2 | 9 |
| 13 | 11 | 2 |
| 17 | 3 | 14 |
| …  | … | … |

In general: `6m ≡ 1 (mod p)` iff `m ≡ 6⁻¹ (mod p)`, and `6m ≡ −1 (mod p)` iff
`m ≡ −6⁻¹ (mod p)`. The mod-5 filter handles p=5 only; every other prime ≡ ±1 (mod 6)
contributes its own pair of forbidden residue classes, none of which are captured by the
sum-closure rule.

The excess elements are composited by 7 most often (since it is the smallest such prime and
has the highest density 2/(7−1) = 1/3), then 11, 13, 17, … in decreasing frequency —
exactly the distribution expected from the prime number theorem. There is no further
algebraic structure: the excess is simply those m reachable by the sum rule that happen
not to be twin-prime cofactors.

````python
def characterise_excess(seeds={1, 2, 3}, N=1000):
    """Show what primes account for the composite 6m±1 in the excess M \\ A002822."""
    from collections import Counter

    sieve = _prime_sieve(6*N + 1)
    a002822 = set(m for m in range(1, N+1) if sieve[6*m-1] and sieve[6*m+1])
    M = generate_from_seeds(seeds, N)
    excess = sorted(M - a002822)

    def smallest_factor(n):
        for p in range(2, int(n**0.5)+1):
            if n % p == 0:
                return p
        return None

    factor_counts = Counter()
    for m in excess:
        for v in (6*m-1, 6*m+1):
            if not sieve[v]:
                f = smallest_factor(v)
                factor_counts[f] += 1

    print(f"Seeds={sorted(seeds)}, N={N}")
    print(f"  |M|={len(M)}, |A002822|={len(a002822)}, |excess|={len(excess)}")
    print()
    print("Smallest prime factor of composite 6m±1 in excess elements:")
    cumulative = 0
    total = sum(factor_counts.values())
    for p, count in sorted(factor_counts.items()):
        cumulative += count
        inv6 = pow(6, -1, p)
        r_minus = inv6 % p        # m ≡ r_minus (mod p) => p | 6m-1
        r_plus  = (-inv6) % p     # m ≡ r_plus  (mod p) => p | 6m+1
        print(f"  p={p:>3}: {count:>4} occurrences  "
              f"(forbidden residues mod {p}: {r_minus} and {r_plus})  "
              f"cumulative: {cumulative/total:.1%}")

characterise_excess(N=10000)
````

# Structure of A067611: Type-0 and Type-1 Elements, and A002822 as Type-2

For each m ≥ 1, classify by the number of primes in {6m−1, 6m+1}:

- **Type 0**: neither 6m−1 nor 6m+1 is prime — both composite. (subset of A067611)
- **Type 1**: exactly one of 6m−1, 6m+1 is prime — a "near-miss". (subset of A067611)
- **Type 2**: both 6m−1 and 6m+1 are prime — a twin prime pair. (A002822)

A067611 is the union of type-0 and type-1; A002822 is exactly type-2.

## Run structure

Consecutive integers form runs of constant type. The run length distributions for
type-0 and type-1 are strikingly similar — both decay roughly geometrically — and the
two types are nearly equally frequent (~52% type-0, ~48% type-1 among A067611 elements).
Type-2 elements (A002822) are sparser and separated by runs of types 0 and 1.

The dominant pattern within A067611 is long stretches of type-1 punctuated by shorter
bursts of type-0, with neither type dominating strongly.

## Factor structure of type-0 elements

For a type-0 element, both 6m−1 and 6m+1 are composite. The most common smallest-factor
pairs are of the form (5, p) or (p, 5) — reflecting that 5 is the most frequent killer
of either neighbour (accounting for two forbidden residue classes mod 5). After 5, the
pairs involve 7, 11, 13, … in decreasing frequency, consistent with the prime number theorem.

````python
def analyse_types(N=10000):
    """Classify all m in [1,N] by number of primes in {6m-1, 6m+1}:
      type 0: both composite  (A067611 subset)
      type 1: exactly one prime (A067611 subset)
      type 2: both prime      (A002822)
    Report frequencies, run length distributions, and factor structure for type-0.
    """
    from itertools import groupby
    from collections import Counter

    sieve = _prime_sieve(6*N + 1)
    typed = [(m, int(sieve[6*m-1]) + int(sieve[6*m+1])) for m in range(1, N+1)]

    counts_by_type = Counter(c for _, c in typed)
    total = len(typed)
    print(f"All m in [1,{N}]: {total} elements")
    for t in (0, 1, 2):
        n = counts_by_type[t]
        label = ["both composite (A067611)", "one prime (A067611)", "both prime (A002822)"][t]
        print(f"  Type {t} ({label}): {n}  ({n/total:.1%})")

    # Run length distributions across all three types
    counts = [c for _, c in typed]
    runs = [(k, len(list(g))) for k, g in groupby(counts)]
    run_lengths = {0: Counter(), 1: Counter(), 2: Counter()}
    for typ, length in runs:
        run_lengths[typ][length] += 1

    max_len = max(max(run_lengths[t]) for t in (0, 1, 2))
    print(f"\nRun length distribution:")
    print(f"{'length':>8}  {'type=0':>10}  {'type=1':>10}  {'type=2':>10}")
    for l in range(1, max_len + 1):
        c0 = run_lengths[0].get(l, 0)
        c1 = run_lengths[1].get(l, 0)
        c2 = run_lengths[2].get(l, 0)
        if c0 or c1 or c2:
            print(f"{l:>8}  {c0:>10}  {c1:>10}  {c2:>10}")

    # Factor pairs for type-0
    def smallest_factor(n):
        for p in range(2, int(n**0.5)+1):
            if n % p == 0:
                return p
        return n

    pair_factors = Counter()
    for m, c in typed:
        if c == 0:
            pair_factors[(smallest_factor(6*m-1), smallest_factor(6*m+1))] += 1

    print(f"\nMost common factor pairs (sf(6m-1), sf(6m+1)) for type-0 elements:")
    for pair, cnt in pair_factors.most_common(12):
        print(f"  {pair}: {cnt}")

analyse_types(N=20000)
````

# Goldbach-Twin Conjecture

**Conjecture 4a** (Dubner, 1996): *Every even number greater than 4208 is the sum of two t-primes.*

A t-prime is a prime with a twin — i.e. a prime p such that p−2 or p+2 is also prime.
This is a strengthening of Goldbach's conjecture requiring both summands to be t-primes.

Note that this conjecture implies the Twin Prime Conjecture: if the set of twin primes were
finite, then for sufficiently large even n no twin prime p could satisfy p ≤ n−2 with n−p
also prime, and the conjecture would fail.

Dubner verified this up to 2×10¹⁰. We verify it here for even numbers up to 20000, also
checking the weaker form (at least one summand a t-prime) which holds from n = 6.

````python
def verify_goldbach_twin(N=10000):
    """Verify Dubner's Conjecture 4a: every even n > 4208 is the sum of two t-primes.
    Also check the weaker form: at least one summand is a t-prime.

    Reports all exceptions to each form and example decompositions.
    """
    sieve = _prime_sieve(2*N + 10)

    twin_primes = sorted(
        p for p in range(2, 2*N + 1)
        if sieve[p] and (
            (p - 2 >= 2 and sieve[p - 2]) or
            (p + 2 <= 2*N + 10 and sieve[p + 2])
        )
    )
    twin_set = set(twin_primes)

    weak_failures = []   # no summand is a t-prime
    strong_failures = [] # not both summands are t-primes

    for n in range(4, 2*N + 1, 2):
        weak_ok = False
        strong_ok = False
        for p in twin_primes:
            if p >= n:
                break
            q = n - p
            if q >= 2 and sieve[q]:
                weak_ok = True
                if q in twin_set:
                    strong_ok = True
                    break
        if not weak_ok:
            weak_failures.append(n)
        if not strong_ok:
            strong_failures.append(n)

    print(f"Checked even numbers from 4 to {2*N}.")
    print(f"t-primes up to {2*N}: {len(twin_primes)}")

    print(f"\nWeak form (≥1 summand a t-prime) failures: {weak_failures if weak_failures else 'none (except n=4)'}")

    strong_above_4208 = [n for n in strong_failures if n > 4208]
    print(f"\nStrong form (both summands t-primes, Dubner Conj. 4a) failures:")
    print(f"  All failures: {strong_failures[:15]}{'...' if len(strong_failures)>15 else ''}")
    print(f"  Failures > 4208: {strong_above_4208[:10] if strong_above_4208 else 'none'}")
    print(f"  Largest failure: {max(strong_failures)}")

    print("\nExample strong-form decompositions:")
    for n in [100, 1000, 4210, 10000, 20000]:
        for p in twin_primes:
            if p >= n: break
            q = n - p
            if q >= 2 and sieve[q] and q in twin_set:
                print(f"  {n} = {p} (t-prime) + {q} (t-prime)")
                break

verify_goldbach_twin(N=10000)
````

````python
import matplotlib.pyplot as plt

def middle_number_pairs(n):
    """For each twin prime center m in A002822(1..n), find all pairs (a, b)
    of smaller twin prime centers such that a + b == m.
    Returns a dict mapping m -> list of (a, b) pairs.
    """
    sieve = _prime_sieve(6*n + 1)
    centers = list(A002822(range(1, n+1), sieve))
    center_set = set(centers)
    result = {}
    for m in centers:
        pairs = []
        for a in centers:
            if a >= m:
                break
            b = m - a
            if b in center_set and b >= a:
                pairs.append((a, b))
        result[m] = pairs
    return result

def plot_xy_for_m(m, n=None):
    """Plot the x-y plane for all tuples with the given twin prime center m,
    and overlay the witness (x,y) coordinates for each member of every
    middle-number decomposition (p, q) where p + q == m.

    For each decomposition, thin arrows are drawn from each summand witness
    to the centroid of m's witness cloud, and lines connect p witnesses to
    q witnesses.

    n defaults to m+1 if not specified.
    Green stars:    c > 0 (witness from m-1 neighbour).
    Red stars:      c < 0 (witness from m+1 neighbour).
    Teal/salmon dots: witnesses for p or q, coloured by their own c value.
    """
    if n is None:
        n = m + 1
    all_tuples = list(a002822_augmented(n))
    tuples_for = {}
    for t in all_tuples:
        tuples_for.setdefault(t["m"], []).append(t)

    tuples = tuples_for.get(m, [])
    if not tuples:
        print(f"No tuples found for m={m} (is {6*m-1},{6*m+1} a twin prime pair?)")
        return

    pairs = middle_number_pairs(n).get(m, [])

    # centroid of m's witness cloud
    cx = sum(t["x"] for t in tuples) / len(tuples)
    cy = sum(t["y"] for t in tuples) / len(tuples)

    fig, ax = plt.subplots(figsize=(6, 6))

    for t in tuples:
        color = "green" if t["c"] > 0 else "red"
        ax.scatter(t["x"], t["y"], color=color, marker="*", s=80, zorder=2)
        ax.annotate(f'({t["x"]},{t["y"]})', (t["x"], t["y"]),
                    textcoords="offset points", xytext=(4, 4), fontsize=7)

    for (p, q) in pairs:
        p_tuples = tuples_for.get(p, [])
        q_tuples = tuples_for.get(q, [])
        for member, member_tuples in ((p, p_tuples), (q, q_tuples)):
            for t in member_tuples:
                color = "teal" if t["c"] > 0 else "salmon"
                ax.scatter(t["x"], t["y"], color=color, marker=".", s=15, zorder=3)
                ax.annotate(f'{member}:({t["x"]},{t["y"]})', (t["x"], t["y"]),
                            textcoords="offset points", xytext=(4, -10), fontsize=7, color=color)
                # ax.annotate("", xy=(cx, cy), xytext=(t["x"], t["y"]),
                #             arrowprops=dict(arrowstyle="-|>", color=color,
                #                            lw=0.5, mutation_scale=6))
        for tp in p_tuples:
            for tq in q_tuples:
                ax.plot([tp["x"], tq["x"]], [tp["y"], tq["y"]],
                        color="blue", lw=0.4, alpha=0.5, zorder=1)

    ax.axhline(0, color="grey", linewidth=0.5)
    ax.axvline(0, color="grey", linewidth=0.5)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_title(f"Witnesses for m={m}  ({6*m-1}, {6*m+1})\n"
                 f"★ green=c>0, red=c<0  ·  teal=summand c>0, salmon=summand c<0")
    ax.legend(handles=[
        plt.Line2D([0],[0], marker='*', color='w', markerfacecolor='green',  markersize=8, label='m: c>0'),
        plt.Line2D([0],[0], marker='*', color='w', markerfacecolor='red',    markersize=8, label='m: c<0'),
        plt.Line2D([0],[0], marker='.', color='w', markerfacecolor='teal',   markersize=8, label='summand: c>0'),
        plt.Line2D([0],[0], marker='.', color='w', markerfacecolor='salmon', markersize=8, label='summand: c<0'),
    ])
    plt.tight_layout()
    plt.show()
````

# Gap Ratio Plot

The plot below shows the ratio `gap(m) / m` for successive elements of A002822, on a logarithmic
scale, where `gap(m) = m - prev(m)`.

## What the conjectures predict

- **Twin Prime Conjecture** (and the Bridging Conjecture): A002822 is infinite, so the plot
  continues indefinitely — it never terminates at a largest point. If the Twin Prime Conjecture
  were false, the plot would have a final rightmost point beyond which no further data exists.

- **The Bridging Conjecture**: Requires that for every K, some pair of elements of A002822 below K
  sums to an element of A002822 above K. If the Bridging Conjecture were false, beyond some
  threshold K no such bridging pair would exist — the sequence might still continue (if the Twin
  Prime Conjecture holds independently) but all sufficiently large elements would be unreachable
  as sums of smaller ones. This would not necessarily be visible in the gap ratio plot directly,
  but would manifest in the decomposition count plot below.

- **Conjecture 1 (Dubner)**: Every middle number greater than 6 has at least one decomposition.
  Conjecture 1 (Dubner) is independent of whether the sequence is infinite — it could hold for a
  finite sequence or fail for an infinite one. Falsity would not be directly visible in the gap
  ratio, but would appear in the decomposition count plot as a zero count for some m > 1.

- **General trend**: If the ratio `gap / m` trends toward zero, elements of A002822 are becoming
  relatively denser, consistent with (but not proving) all three conjectures. If the ratio grows
  without bound, the sequence is thinning out — not in itself a falsification, but tension with
  the additive structure required by Wagler and the Bridging Conjecture.


````python
def plot_gap_ratio(n=10000):
    """Plot the ratio of the gap between successive A002822 elements
    to the second (larger) element, on a logarithmic y-scale.
    The dashed line at 1/2 marks the bound implied by the Bridging Conjecture:
    m_{n+1} < 2*m_n is equivalent to gap/m_{n+1} < 1/2.
    """
    import numpy as np

    sieve = _prime_sieve(6*n + 1)
    centers = sorted(A002822(range(1, n+1), sieve))

    second = centers[1:]
    gaps = [b - a for a, b in zip(centers, second)]
    ratios = [g / b for g, b in zip(gaps, second)]

    # Rolling mean trend line
    window = max(1, len(ratios) // 25)
    trend = np.convolve(ratios, np.ones(window)/window, mode='valid')
    trend_x = [second[i] for i in range(window//2, window//2 + len(trend))]

    fig, ax = plt.subplots(figsize=(10, 4))
    ax.plot(second, ratios, lw=0.4, color='steelblue', alpha=0.6, label='gap / m')
    ax.plot(trend_x, trend, color='navy', lw=1.5, label=f'Rolling mean (window={window})')
    ax.axhline(y=0.5, color='red', linestyle='--', linewidth=1,
               label='Bridging Conjecture bound (gap/m < 1/2)')
    ax.set_yscale('log')
    ax.set_xlabel('m (second element)')
    ax.set_ylabel('gap / m  (log scale)')
    ax.set_title(f'Gap ratio between successive A002822 elements (n={n})')
    ax.legend()
    plt.tight_layout()
    plt.show()

plot_gap_ratio()
````

# Decomposition Count Plot

The plot below shows, for each m in A002822, the number of ways it can be written as a sum of
two smaller elements of A002822.

## What the conjectures predict

- **Conjecture 1 (Dubner)**: Every element of A002822 greater than 1 has at least one decomposition,
  so the count should never reach zero for m > 1. If Conjecture 1 (Dubner) were false, some m > 1
  would have a count of zero. This is the most directly visible signature in this plot: a zero
  for any m > 1 falsifies Conjecture 1 (Dubner) outright.

- **The Bridging Conjecture**: Requires that the count never *permanently* vanishes beyond any
  threshold K — i.e., there is always some m > K with count ≥ 1 and both summands below K. The
  Bridging Conjecture could fail even if Conjecture 1 holds (if A002822 is finite), or could hold even
  if Conjecture 1 fails (if the occasional zero-count elements are isolated rather than permanent).
  Falsity of the Bridging Conjecture would appear as the count becoming permanently zero beyond
  some point, or the sequence terminating.

- **Twin Prime Conjecture**: If A002822 is finite, the plot terminates. A permanently zero count
  beyond the largest element is the signature of the Twin Prime Conjecture being false.

- **General trend**: A count growing roughly proportionally to m is consistent with elements of
  A002822 being distributed similarly to primes (by analogy with the number of Goldbach
  representations of an even number, which grows roughly as m / log²m). A count that grows faster
  than linearly would suggest unusually dense packing.


````python
def plot_decomposition_count(n=10000):
    """Plot the number of ways each A002822 element m can be decomposed
    as a sum of two smaller A002822 elements, as a function of m.
    """
    import numpy as np

    pairs = middle_number_pairs(n)
    ms = sorted(pairs.keys())
    counts = [len(pairs[m]) for m in ms]

    # Rolling mean trend line
    window = max(1, len(ms) // 25)
    trend = np.convolve(counts, np.ones(window)/window, mode='valid')
    trend_x = [ms[i] for i in range(window//2, window//2 + len(trend))]

    # m / log^2(m) reference curve, scaled to match the data
    ms_arr = np.array(ms, dtype=float)
    ref = ms_arr / np.log(ms_arr)**2
    # Scale so the reference curve passes through the rolling mean at its midpoint
    mid = len(trend) // 2
    scale = trend[mid] / (trend_x[mid] / np.log(trend_x[mid])**2) if trend_x[mid] > 1 else 1.0
    ref_scaled = scale * ref

    fig, ax = plt.subplots(figsize=(10, 4))
    ax.plot(ms, counts, lw=0.4, color='purple', alpha=0.5, label='decomposition count')
    ax.plot(trend_x, trend, color='indigo', lw=1.5, label=f'Rolling mean (window={window})')
    ax.plot(ms, ref_scaled, color='orange', lw=1.2, linestyle='--',
            label=r'$c \cdot m / \log^2 m$ (Goldbach analogy)')
    ax.set_xlabel('m')
    ax.set_ylabel('number of decompositions')
    ax.set_title(f'Number of ways to write m as a sum of two A002822 elements (n={n})')
    ax.legend()
    plt.tight_layout()
    plt.show()

plot_decomposition_count()
````

````python
def verify_no_three_consecutive(n=10000):
    """Verify computationally that no three consecutive integers appear in A002822.

    Also shows, for each consecutive pair (m, m+1), which of the six candidate
    primes 6m±1, 6(m+1)±1, 6(m+2)±1 is divisible by 5 — confirming the
    theoretical argument.
    """
    sieve = _prime_sieve(6*n + 1)
    centers = sorted(A002822(range(1, n+1), sieve))
    center_set = set(centers)

    triples = [(m, m+1, m+2) for m in centers if m+1 in center_set and m+2 in center_set]

    print(f"Checked A002822 up to n={n} ({len(centers)} terms).")
    if triples:
        print(f"COUNTEREXAMPLE FOUND: {triples}")
    else:
        print("Confirmed: no three consecutive terms exist.")

    # Demonstrate the divisibility-by-5 argument for consecutive pairs
    print("\nFor each consecutive pair (m, m+1) in A002822, the six candidate values")
    print("and which is divisible by 5:\n")
    pairs_found = 0
    for m in centers:
        if m + 1 in center_set:
            candidates = [
                (f"6·{m}-1",   6*m - 1),
                (f"6·{m}+1",   6*m + 1),
                (f"6·{m+1}-1", 6*(m+1) - 1),
                (f"6·{m+1}+1", 6*(m+1) + 1),
                (f"6·{m+2}-1", 6*(m+2) - 1),
                (f"6·{m+2}+1", 6*(m+2) + 1),
            ]
            div5 = [(label, v) for label, v in candidates if v % 5 == 0]
            print(f"  m={m}, m+1={m+1}: divisible by 5 → {div5}")
            pairs_found += 1
            if pairs_found >= 10:
                print("  ...")
                break

verify_no_three_consecutive()
````
