# The Middle Number Conjecture

In 1996, Stephen Wagler communicated a question to Paulo Ribenboim: what is the smallest
middle number that is not the sum of two t-primes? Ribenboim asked Harvey Dubner to investigate.
Dubner's paper (1999–2000) is the earliest known published treatment; he formulated four
conjectures in response. Wagler's question is closest in spirit to Conjecture 2 (see below).

**Conjecture 1** (Dubner, 1999): *Every middle number greater than 6 is the sum of two middle numbers.*

A "middle number" is a multiple of 6 sandwiched between a pair of twin primes: if m is in A002822
(i.e. 6m−1 and 6m+1 are both prime), then 6m is the corresponding middle number. A002822 records
the cofactor m, not the middle number itself. A t-prime is a prime which has a twin.

Conjecture 1, translated into A002822 terms, states: for every m in A002822 with 6m > 6,
there exist a, b in A002822 such that 6m = 6a + 6b, i.e. m = a + b.

Dubner verified Conjecture 1 up to 2×10¹⁰, finding that the following multiples of 6 are not
sums of two middle numbers: 96, 402, 516, 786, 906, 1116, 1146, 1266, 1356, 3246, 4206.
None of these are middle numbers themselves, so Conjecture 1 (which applies only to middle
numbers) was not contradicted. The largest non-representable multiple of 6 is 4206, leading to:

**Conjecture 3** (Dubner): *Every multiple of 6 greater than 4206 is the sum of two middle numbers.*

Dubner also states two further conjectures involving t-primes directly:

**Conjecture 2** (Dubner): *Every middle number greater than 6 is the sum of two t-primes.*

**Conjecture 4a** (Dubner): *Every even number greater than 4208 is the sum of two t-primes.*

Conjecture 4a is a strengthening of Goldbach's conjecture requiring both summands to be t-primes
(primes with a twin). It is closely related to our Goldbach-Twin Conjecture below.

We compute, for each m in A002822, the list of pairs (a, b) in A002822 such that a + b = m.

# The Bridging Conjecture

**Bridging Conjecture**: *For all K, there exist a, b, k in A002822 such that a < K, b < K, and k = a + b > K.*

That is: no matter how large K is, one can always find two elements of A002822 below K whose sum
is also in A002822 and exceeds K. In middle number terms: there exist middle numbers 6a and 6b,
both below 6K, whose sum 6k = 6a + 6b is also a middle number and exceeds 6K.

*Note: the current author suspects that Wagler, Dubner, or Zwillinger have already stated an
equivalent conjecture, but has not yet located the precise reference.*

## Relationship to Dubner's Conjecture 3

Dubner's paper states:

**Conjecture 3 (Dubner)**: *Every sufficiently large number divisible by 6 is the sum of two middle numbers.*

More precisely: every multiple of 6 greater than 4206 is the sum of two middle numbers
(i.e. two elements of {6m : m ∈ A002822}).

Conjecture 3 (Dubner) is strictly stronger than Conjecture 1 (Dubner), which only requires
middle numbers themselves to decompose — Conjecture 3 (Dubner) requires *all* sufficiently
large multiples of 6 to decompose, whether or not they are middle numbers.

**Conjecture 3 (Dubner) implies the Bridging Conjecture:** Assume Conjecture 3 (Dubner) holds.
Given any K, let k be the smallest element of A002822 greater than K — such k exists assuming
the Twin Prime Conjecture. Since 6k is a multiple of 6 greater than 4206 (for large enough K),
Conjecture 3 (Dubner) guarantees 6k = 6a + 6b for some a, b ∈ A002822, i.e. k = a + b. Since
k is the *smallest* element of A002822 greater than K, any element of A002822 less than k is
≤ K, so a, b ≤ K. Since a + b = k > K they cannot both equal K, so a, b < K. Thus a, b, k
witness the Bridging Conjecture. ∎

## Relationship to Conjecture 1 (Dubner) and the Twin Prime Conjecture

Conjecture 1 (Dubner) and the Twin Prime Conjecture together imply the Bridging Conjecture:

**Twin Primes ∧ Conjecture 1 → Bridging Conjecture:**
Assume both hold. Given any K, since A002822 is infinite (Twin Primes), let k be the smallest
element of A002822 greater than K. By Conjecture 1, k = a + b for some a, b in A002822 with
a, b < k. Since k is the *smallest* element of A002822 greater than K, any element of A002822
less than k must be ≤ K. Therefore a ≤ K and b ≤ K, and since a + b = k > K they cannot both
equal K, so a, b < K. Thus a, b, k satisfy the Bridging Conjecture. ∎

Conversely, the Bridging Conjecture implies the Twin Prime Conjecture:

**Bridging Conjecture → Twin Primes:**
For any K, the Bridging Conjecture gives an element k = a + b of A002822 with k > K. Since K was
arbitrary, A002822 is infinite — i.e. there are infinitely many twin prime pairs. ∎

Therefore, assuming Conjecture 1 (Dubner), the Bridging Conjecture and the Twin Prime
Conjecture are equivalent:

> Conjecture 1 (Dubner) → ( Bridging Conjecture ↔ Twin Prime Conjecture )

Conjecture 1 itself is independent — it could hold for a finite set of elements of A002822
(in which case the Twin Prime Conjecture is false and the Bridging Conjecture is false), or fail
for an infinite one (in which case the Twin Prime Conjecture could still be true while the
Bridging Conjecture remains open).

# Gap Bound and Density Implications of the Bridging Conjecture

## The gap bound

The Bridging Conjecture gives, for any K, elements a, b, k in A002822 with a < K, b < K,
and k = a + b > K. Since a < K and b < K:

```
k = a + b < K + K = 2K
```

So: **for every K, there exists k in A002822 with K < k < 2K**.

This does not require Conjecture 1 — it is a direct consequence of the Bridging
Conjecture's statement alone.

Equivalently: if m_n < m_{n+1} are consecutive elements of A002822, then m_{n+1} < 2·m_n.
No gap between consecutive elements can exceed the smaller element.

## Density lower bound

Since every interval (m, 2m) contains at least one element of A002822, starting from
m_1 = 1 and applying the gap bound inductively:

```
m_1 = 1,  m_2 < 2,  m_3 < 4,  m_4 < 8,  ...,  m_n < 2^(n-1)
```

Inverting: if m_n ≤ x, then n ≥ log₂(x) + 1. So:

```
π₂(x) = |{m ∈ A002822 : m ≤ x}| ≥ ⌊log₂(x)⌋
```

This is a conditional lower bound: **assuming the Bridging Conjecture, π₂(x) ≫ log x**.

This is far weaker than the conjectured truth (roughly x / log²x by Hardy–Littlewood
heuristics), but it is a rigorous consequence of the Bridging Conjecture alone, with no
appeal to analytic number theory.

## What Conjecture 1 adds

The gap bound and density lower bound follow from the Bridging Conjecture alone. Conjecture 1
(Dubner) says every element of A002822 beyond 1 decomposes as a sum of two smaller elements —
not just the specific k produced by the Bridging Conjecture, but all of them. Note that sums of pairs from A002822 need not land in A002822 — many land in A067611 instead.
What Conjecture 1 asserts is that every element of A002822 beyond 1 is reachable as a sum of
two smaller elements of A002822; no element is an additive dead-end. This is a non-trivial
demand on the self-sumset structure of A002822, and is a stronger statement than the Bridging
Conjecture alone.

# No three consecutive terms in A002822

**Observation** (Jonathan Seymour, May 14 2026):

No three terms of A002822 greater than 3 are consecutive. Otherwise there would exist six integers
6m−1, 6m+1, 6m+5, 6m+7, 6m+11, 6m+13 that are all prime. One of these must be
divisible by 5, which is impossible for primes greater than 5. (The exception {1, 2, 3} corresponds
to the primes {5, 7, 11, 13, 17, 19}, where 5 itself is prime rather than composite.)

Hence every term of A002822 greater than 3 is necessarily of the form 6ab ± a ± b ± 1. This condition
is necessary but not sufficient for membership in the sequence.

# Goldbach-Twin Conjecture

**Conjecture 4a** (Dubner, 1996): *Every even number greater than 4208 is the sum of two t-primes.*

A t-prime is a prime with a twin — i.e. a prime p such that p−2 or p+2 is also prime.
This is a strengthening of Goldbach's conjecture requiring both summands to be t-primes.

Note that this conjecture implies the Twin Prime Conjecture: if the set of twin primes were
finite, then for sufficiently large even n no twin prime p could satisfy p ≤ n−2 with n−p
also prime, and the conjecture would fail.

Dubner verified this up to 2×10¹⁰. We verify it here for even numbers up to 20000, also
checking the weaker form (at least one summand a t-prime) which holds from n = 6.

# Gap Ratio Plot

The plot below shows the ratio `gap(m) / m` for successive elements of A002822, on a logarithmic
scale, where `gap(m) = m - prev(m)`.

## What the conjectures predict

- **Twin Prime Conjecture** (and the Bridging Conjecture): A002822 is infinite, so the plot
  continues indefinitely — it never terminates at a largest point. If the Twin Prime Conjecture
  were false, the plot would have a final rightmost point beyond which no further data exists.

- **The Bridging Conjecture**: Requires that for every K, some pair of elements of A002822 below K
  sums to an element of A002822 above K. If the Bridging Conjecture were false, beyond some
  threshold K no such bridging pair would exist — the sequence might still continue (if the Twin
  Prime Conjecture holds independently) but all sufficiently large elements would be unreachable
  as sums of smaller ones. This would not necessarily be visible in the gap ratio plot directly,
  but would manifest in the decomposition count plot below.

- **Conjecture 1 (Dubner)**: Every middle number greater than 6 has at least one decomposition.
  Conjecture 1 (Dubner) is independent of whether the sequence is infinite — it could hold for a
  finite sequence or fail for an infinite one. Falsity would not be directly visible in the gap
  ratio, but would appear in the decomposition count plot as a zero count for some m > 1.

- **General trend**: If the ratio `gap / m` trends toward zero, elements of A002822 are becoming
  relatively denser, consistent with (but not proving) all three conjectures. If the ratio grows
  without bound, the sequence is thinning out — not in itself a falsification, but tension with
  the additive structure required by Wagler and the Bridging Conjecture.

# Decomposition Count Plot

The plot below shows, for each m in A002822, the number of ways it can be written as a sum of
two smaller elements of A002822.

## What the conjectures predict

- **Conjecture 1 (Dubner)**: Every element of A002822 greater than 1 has at least one decomposition,
  so the count should never reach zero for m > 1. If Conjecture 1 (Dubner) were false, some m > 1
  would have a count of zero. This is the most directly visible signature in this plot: a zero
  for any m > 1 falsifies Conjecture 1 (Dubner) outright.

- **The Bridging Conjecture**: Requires that the count never *permanently* vanishes beyond any
  threshold K — i.e., there is always some m > K with count ≥ 1 and both summands below K. The
  Bridging Conjecture could fail even if Conjecture 1 holds (if A002822 is finite), or could hold even
  if Conjecture 1 fails (if the occasional zero-count elements are isolated rather than permanent).
  Falsity of the Bridging Conjecture would appear as the count becoming permanently zero beyond
  some point, or the sequence terminating.

- **Twin Prime Conjecture**: If A002822 is finite, the plot terminates. A permanently zero count
  beyond the largest element is the signature of the Twin Prime Conjecture being false.

- **General trend**: A count growing roughly proportionally to m is consistent with elements of
  A002822 being distributed similarly to primes (by analogy with the number of Goldbach
  representations of an even number, which grows roughly as m / log²m). A count that grows faster
  than linearly would suggest unusually dense packing.

````python
def plot_decomposition_gap_ratio(n=5000):
    """
    For each m in A002822 union D, plot g/m for every decomposition a+b=m (g=b-a),
    on a log-log scale. A243956 elements use the trivial decomposition (0, m) giving g/m=1.
    For A002822 elements, all decompositions are shown; the largest-gap one is
    highlighted in orange.
    """
    D = {1, 16, 67, 86, 131, 151, 186, 191, 211, 226, 541, 701}

    sieve = prime_sieve(6 * n + 1)
    a002822 = sorted(m for m in range(1, n+1) if sieve[6*m-1] and sieve[6*m+1])
    a002822_set = set(a002822)
    pairs = middle_number_pairs(n)

    xs_inner, ys_inner = [], []
    xs_largest, ys_largest = [], []
    xs_D, ys_D = [], []

    for m in a002822:
        decomps = pairs.get(m, [])
        if not decomps:
            xs_D.append(m)
            ys_D.append(1.0)
            continue
        gaps = [(b - a, a, b) for (a, b) in decomps]
        max_gap = max(g for g, _, _ in gaps)
        for g, a, b in gaps:
            ratio = g / m
            if g == max_gap:
                xs_largest.append(m)
                ys_largest.append(ratio)
            else:
                xs_inner.append(m)
                ys_inner.append(ratio)

    for m in D:
        if m not in a002822_set:
            xs_D.append(m)
            ys_D.append(1.0)

    fig, ax = plt.subplots(figsize=(10, 5))
    ax.scatter(xs_inner, ys_inner, s=1, alpha=0.3, color='steelblue',
               label='Decompositions (non-largest)')
    ax.scatter(xs_largest, ys_largest, s=4, alpha=0.6, color='darkorange',
               label='Largest-gap decomposition')
    ax.scatter(xs_D, ys_D, s=12, alpha=0.8, color='red', marker='x',
               label='A243956 (trivial: g/m = 1)')
    ax.axhline(y=1.0, color='red', linestyle='--', linewidth=0.8, alpha=0.5)
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel('m (log scale)')
    ax.set_ylabel('g / m  (log scale),  g = b − a')
    ax.set_title(f'Gap ratio g/m for all decompositions of A002822 ∪ A243956  (n={n})')
    ax.legend(fontsize=9, markerscale=3)
    plt.tight_layout()
    plt.show()

plot_decomposition_gap_ratio(n=5000)
````

# Decomposition Gap Ratio Plot

For each m in A002822 ∪ D, we plot g/m for every decomposition a+b=m, where g = b−a:

- **A002822 elements**: all decompositions a+b=m with a,b ∈ A002822, a ≤ b are shown.
  The decomposition with the largest gap is highlighted in orange; the rest are blue.
- **A243956 elements**: no decomposition exists within A002822, so the trivial
  decomposition (0, m) is used, giving g/m = 1 always. These are shown as red crosses.

Both axes are logarithmic. The dashed line at g/m = 1 marks the trivial upper bound
(g = b−a ≤ b+a = m, with equality only when a = 0).

## What to look for

- The A243956 elements cluster at small m and are pinned at g/m = 1 — they have no
  non-trivial decomposition, so the gap is as large as it can be.
- The orange upper envelope (largest-gap decomposition per m) descends from near 1
  at small m, reflecting that early A002822 elements have few decompositions and
  those decompositions are lopsided. As m grows, more balanced decompositions appear
  and the largest gap becomes a smaller fraction of m.
- The blue cloud fans downward on the log-log scale, showing the growing spread of
  decompositions as m increases — consistent with the decomposition count growing
  like m/(log m)².
